# Chi-Square

```python
from scipy.stats import chi2_contingency as chi
chi_corr_cat_cols=[]
chi_non_corr_cols=[]
def chi_square(df,out_column):
    cat_cols=[col for col in df.columns if df[col].dtype=='object']
    for col in cat_cols:
        contingency_table=pd.crosstab(df[col],df[out_column])
        p_value=chi(contingency_table)[1]
        if p_value<0.05:
            chi_corr_cat_cols.append(col)
        else:
            chi_non_corr_cols.append(col)
            
    
chi_square(ic_df,'fraud_reported')

print(chi_corr_cat_cols)
```

```python
print(chi_non_corr_cols)
```

# Anova

```python
cont_cols=[col for col in ic_df.columns if ic_df[col].dtype=='int64' or ic_df[col].dtype=='float64']
from scipy.stats import f_oneway
anova_corr_cols=[]
anova_non_corr_cols=[]
for col in cont_cols:
    
    y_list=list(ic_df[ic_df['fraud_reported']=='Y'][col])
    n_list=list(ic_df[ic_df['fraud_reported']=='N'][col])
    p_value=f_oneway(y_list,n_list)[1]
    if p_value<0.05:
        anova_corr_cols.append(col)
    else:
        anova_non_corr_cols.append(col)
print(anova_corr_cols)
print(anova_non_corr_cols)
```

```python
Total_non_corr=chi_non_corr_cols+anova_non_corr_cols
```

```python
#identifying the columns tp drop
cols_to_drop=[]
for col in ic_df.columns:
    if ic_df[col].nunique()==1 or ic_df[col].nunique()==ic_df.shape[0]:
        cols_to_drop.append(col)
        
cols_to_drop+=Total_non_corr
cols_to_drop.append('fraud_reported')
cols_to_drop
```

```python
ic_df.drop(columns=cols_to_drop,inplace=True)# dropping all the identified columns(from the above steps
```
